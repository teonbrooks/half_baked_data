This notebook is a quick demonstration of how you can load data, aggregate it, and display it.

```python
# This imports a library that can read data files.
import pandas as pd
# This imports seaborn, a data visualization library.
import seaborn as sns
```

```python
# This reads a tab-separated value (TSV) file and saves it to a dataframe, an excel-like object
df = pd.read_csv("gapminder.tsv", sep="\t")
```

```python
df
```

```python
# This function returns the first five elements
df.head()
```

```python
df.describe()
```

```python
# This creates a lineplot that show the life expectancy over time
sns.lineplot(x="year", y="lifeExp", data=df)
```

Let's see if we can look at this data in more detail

```python
# This creates a lineplot that show the life expectancy over time but now we are going to separate by country
sns.lineplot(x="year", y="lifeExp", hue="continent", data=df)
```

```python
# Now let's do the same for population
sns.lineplot(x="year", y="pop", hue="continent", data=df)
```

```python
# This creates a lineplot that show the population over time
sns.relplot(x="year", y="pop", col="continent", data=df, kind="line")
```

This is kinda boring looking. Let's add some color to the individual plots.

```python
sns.relplot(x="year", y="pop", col="continent", data=df, kind="line", hue="continent", palette="pastel")
```

## Selecting a new countries

We've been doing all our grouping at the continent level but maybe we want to see what the trends are at a country level. Let's see what countries are listed in this dataset.

```python
# This returns all the unique countries in our dataset
df.country.unique()
```

```python
# Let's select a few countries to look at side-by-side
small_df = df.query("country in ['United States', 'Brazil', 'Mexico', 'France']")
```

```python
# Let's preview our new dataframe.
small_df
```

```python
sns.relplot(x="year", y="lifeExp", col="country", data=small_df, kind="line")
```

```python
sns.relplot(x="year", y="lifeExp", col="country", data=small_df, kind="line", hue="country", palette="dark")
```
